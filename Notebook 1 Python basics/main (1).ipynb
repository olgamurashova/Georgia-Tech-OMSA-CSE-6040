### Exercise 7: (2 points)
**count_word_lengths_test**  

**Your task:** define `count_word_lengths_test` as follows:


Write a function that returns a list containing the lengths of words in a string

**Inputs**:
- `s` (str): a phrase of words (lowercase alphabetic characters) separated by multiple character spaces

**Return**:
- `word_lengths` (list): Contains the lengths of words from input. If input is empty, function should return an empty list.


````python
### Solution - Exercise 7  
def count_word_lengths_test(s: str) -> list:
    ###
    ### YOUR CODE HERE
    ###
    word_lengths=[len(word) for word in s.split()]
    return word_lengths

### Demo function call
s='the quick  brown   fox jumped over     the lazy  dog'
print(f'result={count_word_lengths_test(s)}')
````
